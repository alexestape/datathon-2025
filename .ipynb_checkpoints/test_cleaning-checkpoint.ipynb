```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("test.csv", sep=";")
```

```python
df = df1.copy(deep=False)
```

```python
# adding attribute num_week_iso


df["phase_in"] = pd.to_datetime(df["phase_in"], dayfirst=True)
df["phase_out"]   = pd.to_datetime(df["phase_out"], dayfirst=True)

def expand_weeks(row):
    # all days between start and end
    days = pd.date_range(row["phase_in"], row["phase_out"], freq="D")
    # unique ISO week numbers in that span
    weeks = pd.Series(days.isocalendar().week).drop_duplicates().tolist()
    
    # copy all other columns (except the dates)
    base = row.drop(labels=["phase_in", "phase_out"]).to_dict()
    
    # repeat row for each week and add num_week_iso
    out = pd.DataFrame([base] * len(weeks))
    out["num_week_iso"] = weeks
    return out

result = pd.concat(df.apply(expand_weeks, axis=1).to_list(), ignore_index=True)


result

df2 = result.copy(deep=False)


```

```python
df = df2.copy(deep=False) # checkpoint
```

```python
print(df.dtypes) # these are all columns
```

```python


cols_useless = [
'color_rgb',
'image_embedding',
'life_cycle_length',
]


df = df.drop(cols_useless, axis=1)


cols_to_match_train = [
    'aggregated_family',
    'category',
    'waist_type',
    'neck_lapel_type',
    'sleeve_length_type',
    'heel_shape_type',
    'toecap_type',
    'woven_structure',
    'knit_structure',
    'archetype',
    'has_plus_sizes',
    'Unnamed: 28',
    'Unnamed: 29',
    'Unnamed: 30',
    'Unnamed: 31',
    'Unnamed: 32',
]


df = df.drop(cols_to_match_train, axis=1)


df = df.loc[:, df.isna().mean() < 0.90] # only keep attributes that have less than 20% NaNs
```

```python
print(df.dtypes)

df = df.astype({c: "float64" for c in df.select_dtypes("int64").columns})
cols = ["ID", "id_season", "num_week_iso"]
df[cols] = df[cols].astype("object")

print(df.dtypes)
```

```python
df.to_csv("test_ready.csv", index=False)
```

```python
nan_counts = df.isna().sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
nan_counts.plot(kind='bar')
plt.title("NaNs per column")
plt.ylabel("count")
plt.tight_layout()
plt.show()


# without plot:

nan_counts = df.isna().sum()

print(nan_counts)
```

```python
# we replace nans with mode in train.csv

df3 = pd.read_csv("train_ready.csv", sep=",")

# get modes as a list
modes = df3[['length_type','silhouette_type','print_type']].mode().iloc[0].tolist()

# fill df with those modes
df[['length_type','silhouette_type','print_type']] = df[['length_type','silhouette_type','print_type']].fillna(
    dict(zip(['length_type','silhouette_type','print_type'], modes))
)
```

```python
nan_counts = df.isna().sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
nan_counts.plot(kind='bar')
plt.title("NaNs per column")
plt.ylabel("count")
plt.tight_layout()
plt.show()


# without plot:

nan_counts = df.isna().sum()

print(nan_counts)
```
